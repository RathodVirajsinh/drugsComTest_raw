```python
a1=b.groupby('drugName')['uniqueID'].count().sort_values(ascending=False).head(15)
a1.plot.bar()
print(a[["drugName"]].value_counts().head(15))
```

```python

top_drugs = a['drugName'].value_counts().head(15).index
drugs_counts = a['drugName'].value_counts().head(15)


avg_rating = a[a['drugName'].isin(top_drugs)].groupby('drugName')['rating'].mean().reindex(top_drugs)


count_and_rating = pd.concat([drugs_counts,avg_rating],axis=1)
print(count_and_rating.rename(columns={'count':'drugs_Counts','rating':'avg_rating'}))


fig, ax1 = plt.subplots(figsize=(12, 8))


color="lightgreen"
ax1.set_xlabel('count')
ax1.set_ylabel('rating', color='#19aade')
ax1.bar(drugs_counts.index,drugs_counts.values, color=color)
ax1.tick_params(axis='y', labelcolor='#19aade')
ax1.set_xticks(range(len(drugs_counts.index))) 
ax1.set_xticklabels(drugs_counts.index, rotation=45, ha='right')



ax2 = ax1.twinx()
color = '#142459'
ax2.set_ylabel('avg_rating', color='#213B91')
ax2.plot(avg_rating.index, avg_rating.values, color=color, marker='o',markerfacecolor='r')
ax2.tick_params(axis='y', labelcolor='#213B91')


plt.title('Top 20 drugs by Popularity')
fig.tight_layout()


plt.show()
```

```python
avg_condition = a.groupby('condition')['rating'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=avg_condition.values, y=avg_condition.index, palette='magma')
plt.title('Top 10 Conditions by Average Rating')
plt.xlabel('Average Rating')
plt.ylabel('Condition')
plt.show()

```

```python
a['date'] = pd.to_datetime(a['date'])
reviews_by_date = a.groupby(a['date'].dt.to_period('M')).size()

plt.figure(figsize=(10,5))
reviews_by_date.plot(kind='line')
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show()

```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='rating', y='usefulCount', data=a, alpha=0.6)
plt.title('Useful Count vs Rating')
plt.xlabel('Rating')
plt.ylabel('Useful Count')
plt.show()
```
